### Предобработка данных

Напишем функцию, которая будет выводить наглядную диаграмму с процентом пропущенных значений в столбцах датафрейма, где таковые встречаются:

```python
def nan_value_barh(df):
    (
        (df.isna().mean()*100)
        .to_frame()
        .rename(columns = {0:'space'})
        .query('space > 0')
        .sort_values(by = 'space', ascending = True)
        .plot(kind= 'barh', figsize=(19,6), legend = False, fontsize = 16)
        .set_title('Проценты', fontsize = 22, color = 'SteelBlue')    
    )
```

```python
nan_value_barh(data)
```

Приведём названия столбцов к змеиному регистру:

```python
data = data.rename(columns={'cityCenters_nearest': 'city_centers_nearest', 
                            'parks_around3000': 'parks_around_3000', 'ponds_around3000': 'ponds_around_3000'})
```

#### Изучим столбец `total_images`:

Количество картинок целое, нет пропущенных значений. Посмотрим на общую информацию о значениях в столбце:

```python
data['total_images'].describe().to_frame()
```

Аномально высоких значений нет. Нулевые значения вполне себе могут быть (человек мог не загрузить фото по разным причинам; хотя быть может мог быть сбой в системе, из-за чего фото не загрузились). Не будем ничего менять.

#### Изучим столбец `last_price`:

Тип данных столбца `last_price` -- float. Проверим, есть ли там дробные значения:

```python
display(data[data['last_price'] % 1 != 0])
```

Датасет оказался пуст. Можно смело привести все значения к типу int:

```python
data['last_price'] = data['last_price'].astype(int)
```

Посмотрим на общую информацию о значениях в столбце:

```python
data['last_price'].describe().to_frame()
```

Аномальных значений с точки зрения здравого смысла не наблюдается. Квартиры могут стоить очень дорого. Однако нас интересуют не выходящие за рамки обычного показатели. Выведем на экран отсортированные по убыванию значения, которые лежат выше верхнего уса диаграммы размаха:

```python
Q1 = data['last_price'].quantile(0.25)
Q3 = data['last_price'].quantile(0.75)
IQR = Q3 - Q1
data[data['last_price'] > Q3 + 1.5*IQR]['last_price'].sort_values(ascending=False).to_frame()
```

Получилось 2034 значения. Это достаточно много по сравнению с общим объёмом выборки. Посчитаем количество квартир стоимость больше 30 000 000:

```python
data[data['last_price'] > 30000000]['last_price'].count()
```

Таких значений не много. Можно их исключить:

```python
data = data[data['last_price'] <= 30000000]
```

```python
data['total_area'].describe().to_frame()
```

Тем не менее, попробуем оценить долю квартир на рынке с аномально большой площадью. Выведем на экран отсортированные по убыванию значения, которые лежат выше верхнего уса диаграммы размаха:

```python
Q1 = data['total_area'].quantile(0.25)
Q3 = data['total_area'].quantile(0.75)
IQR = Q3 - Q1
data[data['total_area'] > Q3 + 1.5*IQR]['total_area'].sort_values(ascending=False).to_frame()
```

Их количество -- 1065. Это довольно много для исключения. Посчитаем число квартир с площадью больше 190м2:

```python
data['first_day_exposition'] = data['first_day_exposition'].dt.round('1D')
```

Пропущенных значений в столбце нет. Больше ничего менять не будем.

#### Изучим столбец `ceiling_height`

В столбце `ceiling_height` насчитывается большой процент пропусков по сравнению с общим количеством строк:

```python
print(round(data['ceiling_height'].isna().sum()/ data['last_price'].count() * 100, 1), '%')
```

Возможной причиной пропусков могла стать невнимательность или пренебрежение этим параметром собственником. Другой причиной могла быть человеческая лень: не каждый помнит точную высоту потолков в своей квартире, а чтобы её посчитать нужно взять рулетку, залезть под потолок и аккуратно сделать замеры -- слишком много действий для такого казалось бы маловажного параметра.

Мы не можем выкинуть так много строк с пропусками. Кроме того, высота потолков не явлется первостепенным критерием при выборе квартиры. Поэтому, пропущенные значения можно заменить на среднее значение по всей выборке или медиану. Прежде чем заменять, убедимся, что в столбце нет выбросов:

```python
data['ceiling_height'].describe().to_frame()
```

Выбросы есть. Построим диаграмму размаха:

```python
plt.ylim(0, 4)
plt.xlim(0, 2) 
data.boxplot(column='ceiling_height');
```

Выше верхнего уса лежит большое количество выбросов.

```python
Q1 = data['ceiling_height'].quantile(0.25)
Q3 = data['ceiling_height'].quantile(0.75)
IQR = Q3 - Q1
display(data[data['ceiling_height'] < Q1 - 1.5*IQR]['ceiling_height'].sort_values(ascending=False).to_frame())
```

Это аномально низкие значения и их немного. Логичной замены предложить нельзя: если пользователь ошибся проставлением точки, то предполагаемая высота станет больше 10 метров в каждом случае -- это уже выбивается за верхний ус диаграммы размаха. Выкинем аномально низкие значения позже.

Посмотрим на значения, которые лежат выше верхнего уса:

```python
data[data['ceiling_height'] > Q3 + 1.5*IQR]['ceiling_height'].sort_values().to_frame()
```

Их количество:

```python
data[data['ceiling_height'] > Q3 + 1.5*IQR]['ceiling_height'].count()
```

Аномально высоких значений много -- мы не можем просто так их выкинуть. Можно предположить, что пользователь ошибся проставлением точки. Поделим все аномально высокие значения на 10, а затем отбросим все строки, что выйдут за пределы усов диаграммы размаха, оставив пропущенные значения:

```python
data.loc[data['ceiling_height'] > Q3 + 1.5*IQR, 'ceiling_height'] = \
                data[data['ceiling_height'] > Q3 + 1.5*IQR]['ceiling_height'] / 10
data = data[(data['ceiling_height'] <= Q3 + 1.5*IQR) & 
                (data['ceiling_height'] >= Q1 - 1.5*IQR) | data['ceiling_height'].isna()]
```

```python
#создадим датафрейм df с теми строками, где локация -- одна из 10 из условия:
l = data.groupby('locality_name')['one_square_price'].count().sort_values(ascending=False).head(10).index
df = data[data['locality_name'].isin(l)].copy()
#выведем в порядке убывания средней цены таблицу с названием населённого пункта
#и стоимости жилья в нём:
df = df.groupby('locality_name')['one_square_price'].mean().sort_values(ascending=False)

display(df.to_frame().rename(columns={'one_square_price': 'Средняя стоимость квадратного метра млн.'})\
        .rename_axis('Город').apply(lambda x: round(x*1e-06, 2)))

print()
print('Среди 10 населённых пунктов с самым большим количеством объявлений:')
print()
print(f'самая высокая стоимость квадратного метра в населённом пункте {list(df.index)[0]};')
print(f'самая низкая стоимость квадратного метра в населённом пункте {list(df.index)[-1]};')
```
